```
import numpy as np
import pandas as pd
```

```
df = pd.read_csv('TI-CPI 2012-present.csv')
df.head()
```

```
df.drop(columns=['note'], inplace=True)
df.head()
```

```
df.drop(columns=['Indicator name', 'indicator','Country name'], inplace=True)
df
```

```
df
```

```
cpi_yearly = df.pivot(index='country', columns='time',values='value')

cpi_yearly
```

```
cpi_yearly.to_excel("TI_CPI_2012_2018_pivoted.xlsx")
```
