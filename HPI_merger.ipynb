```
import numpy as np
import pandas as pd
```

```
hpi6 = pd.read_excel('trim_hpi_2006.xlsx')
hpi6.head()
```

```
hpi6.drop(columns=['code'], inplace=True)
hpi6.head()
```

```
hpi9 = pd.read_excel('trim_hpi_2009.xlsx')
hpi9.head()
```

```
hpi9.drop(columns=['code'], inplace=True)
hpi9.head()
```

```
hpi12 = pd.read_excel('trim_hpi_2012.xlsx')
hpi12.head()
```

```
hpi16 = pd.read_excel('trim_hpi_2016.xlsx')
hpi16.head()
```

```
hpi_6_9 = pd.merge(hpi6, hpi9,
                    on='country', how='outer',
                   indicator=True)
hpi_6_9
```

```
hpi_6_9.drop(columns=['_merge'], inplace=True)
hpi_6_9.head()
```

```
hpi_6_9.rename(columns={'country':'Country', 'HPI_x':'2006','HPI_y':'2009'}, inplace=True)
hpi_6_9.head()
```

```
hpi_6_12 = pd.merge(hpi_6_9, hpi12,
                    on='Country', how='outer',
                   indicator=True)
hpi_6_12
```

```
hpi_6_12.drop(columns=['_merge'], inplace=True)
hpi_6_12.rename(columns={'Happy Planet Index':'2012'}, inplace=True)
hpi_6_12.head()
```

```
hpi_6_16 = pd.merge(hpi_6_12, hpi16,
                    on='Country', how='outer',
                   indicator=True)
hpi_6_16
```

```
hpi_6_16.drop(columns=['_merge'], inplace=True)
hpi_6_16.rename(columns={'Happy Planet Index':'2016'}, inplace=True)
hpi_6_16.head()
```

```
hpi_6_16.to_excel("HPI_2006_2016.xlsx")
```
