```
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```
rev_gdp = pd.read_excel('CW Rev per GDP NANs.xlsx')
rev_gdp
```

```
cpi = pd.read_excel('PTI CPI NANs.xlsx')
cpi.head()
```

```
hpi = pd.read_excel('HPI TARG FEAT NANs.xlsx')
hpi.head()
```

```
scaler = StandardScaler()
rev_gdp_std = scaler.fit_transform(hpi.iloc[:,1:])
```

```
rev_gdp_std
```

```
slc_rev_gdp_std = pd.DataFrame(rev_gdp_std)
slc_rev_gdp_std.head()
```

```
df1 = pd.DataFrame(hpi.iloc[:,0:1])
df1.head()
```

```
df_rev_gdp_std = pd.concat([df1, slc_rev_gdp_std],axis=1)
df_rev_gdp_std.columns = [hpi.iloc[0:1,:]]
df_rev_gdp_std
```

```
df_rev_gdp_std.to_excel("HPI_std.xlsx")
```
