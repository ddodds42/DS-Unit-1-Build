```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import r2_score

%matplotlib inline

df = pd.read_excel('PWGI Voice Acct NANs.xlsx')
df.head()
```

```
prestrip = df.T

prestrip.head()
```

```
preimputable = prestrip.iloc[2:,:]
preimputable.head()
```

```
imputable = preimputable
```

```
imputable.index = pd.to_datetime(preimputable.index, format='%Y')
imputable.head()
```

```
imputed = imputable.astype(float).interpolate(method='time', limit_direction="both")
imputed.head()
```

```
mergeable = imputed.T

mergeable.head()
```

```
# mergeable.describe()
mergeable.iloc[:,0:1].describe()
```

```
downloadable = pd.concat([df.iloc[:,0:1],mergeable], axis=1)
downloadable
```

```
downloadable.isna().sum()
```

```
downloadable.to_excel("PWGI_Voice_Acct_imp.xlsx")
```
