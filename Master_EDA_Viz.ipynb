```
import numpy as np
import pandas as pd
import plotly.express as px
```

### **Below is the raw csv dataset, open for anyone with this colab notebook to access, so it can run from head to toe without the need to download the dataset from your local machine!**

```
df = pd.read_csv('https://raw.githubusercontent.com/ddodds42/DS-Unit-1-Build/master/Master13.csv', sep=',', encoding='latin-1')
df.head(20)
```

### **This correlation matrix is trimmed to display which attributes are most predictive of the target feature, the Happy Planet Index growth rate and / or mean.**

```
correlations = df.corr()
correlations.iloc[:,:2]
```

###**The strongest correlation was a measly 02.7, plotted below. This cell preceeding the plot trims null values out of the 2 correlated features so plotly can display a trend line with it .**

```
plottable = df.loc[:,['TARG_HPI_trend_coeff','PW_Women_Legis_mean']].dropna()

plottable.isna().sum()
```

```
fig = px.scatter(plottable, x="PW_Women_Legis_mean", y="TARG_HPI_trend_coeff", trendline="ols")
fig.update_layout(
    title_text = 'Most Notable Predictor of National Well-Being: Women in the Legislature')
fig.show()
```

###**The closest correlate to national prosperity was the percentage of women in the national legislative body. Such a finding deserves a stronger correlation than 0.27. Anyhow, the next cell slices and sorts the dataframe in preparation for a heatmap which visualizes the 10-year trend of the Happy Planet Index for each nation. The sort was not necesssary for the heatmap, it just allows us to see which nations were most improving or most declining on the HPI. Many questionable surprises are in this sort, which concludes that Zimbabwe is improving more than Canada, for instance. Though it may be true in the case of Zimbabwe's room for improvement in terms of national prosperity, results like that should lead us to doubt that this metric has predictive validity for which nations are likely to fail or prosper in the long-run.**

```
pd.set_option('display.max_rows', 300)
growth = df.sort_values(by='TARG_HPI_trend_coeff', ascending=False).iloc[:,:3]
growth
```

## Plotly is love. Plotly is life.
**Go ahead and hover, zoom, and pan over the heatmap to see the HPI trend data in each nation. Yellow is fastest growing, dark blue is fastest declining over the decade from 2006-2016 **

```
fig = px.choropleth(df, locations="NS_code",
                    color = "TARG_HPI_trend_coeff",
                    hover_name = "Nation_State",
                    color_continuous_scale = px.colors.sequential.Plasma,
                    width = 2000, height=1000)
fig.update_layout(
    title_text = 'Growth Trend Coefficient on the Happy planet Index, 2006-2016')

fig.show()
```
